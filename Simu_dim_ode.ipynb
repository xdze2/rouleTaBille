```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

```python
# Bille:
#rho = 2500 # kg/m3 (verre) 
rho = 8000 # kg/m3 (acier) 

R = 3e-3 # m, rayon bille
m = 4.0/3.0*np.pi*R**3*rho # masse, kg


omega = 2.0*np.pi*freq*45.0/60. # rad.s


alpha = 0.0000000 # friction


aw = +1.0/7.0  # fraction pour la vitesse de départ


Rstart = 1e-2*8  # rayon de départ en m
y0 = [Rstart, .0, .0, aw*omega*Rstart]  #m, m/s : rX rY   rpX rpY


t1 = 2*np.pi/(omega*2.0/7.0)*.9 # s
dt = 0.005

print 'm:', m*1000, ' gramme'
print 'R:', R*100, ' cm'

g = 9.81


    
def dfdt(t, y, m, R, Omega, alpha ):
    rX, rY = y[0], y[1]
    rX_point, rY_point = y[2], y[3]
    
    Fx, Fy = 0, 0 # Fext(rX, rY)
    
    Jstar = 2.0/5.0  # moment d'inertie, balle pleine
    J = Jstar*m*R**2
    epsilon = Jstar/( Jstar + 1.0 )

    A = -alpha/J*epsilon 
    B = epsilon*Omega
    C = -A*Omega
    D = 1/m/(Jstar+1)
    
    rX_pointpoint = +A*rX_point -B*rY_point -C*rY + D*Fx
    rY_pointpoint = +A*rY_point +B*rX_point +C*rX + D*Fy
    
    return [rX_point, rY_point, rX_pointpoint, rY_pointpoint]



eq = ode(dfdt)
eq.set_initial_value(y0, 0).set_f_params( m, R, omega, alpha )

# In[297]:

allX = []
allY = []
while eq.successful() and eq.t < t1:
    eq.integrate(eq.t+dt)
    allX.append( eq.y[0] )
    allY.append( eq.y[1] )

allX, allY = np.array(allX), np.array(allY)
print 'nombre de pas: ', len(allX)

title = 'm=%0.2fg; R=%0.2fmm; f=%0.2fHz; alpha=%fk' % ( 1e3*m, 1e3*R, Omega/2.0/np.pi, 1e3*alpha )
print title

plt.figure(figsize=(7,7))
ax=plt.subplot(aspect='equal')

ax.add_artist( plt.Circle((100*allX[0], 100*allY[0]), 100*R, alpha=0.2) )

ax.plot( 100*allX, 100*allY )
ax.plot( 100*allX[0], 100*allY[0], 'o' )


ax.plot( 0, 0, 'o' )
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.title(title)
#plt.show()

# Dans le referenciel tournant:
time = np.array( range(len(allX))) *dt

anglesRef = (time*omega - np.pi  ) % (2*np.pi) - np.pi

angles = np.arctan2( allY, allX)
rho = np.sqrt(allY**2 + allX**2  )
 
#plt.figure(figsize=(7,7))
#plt.plot(time, angles-anglesRef )

Xtournant = rho * np.cos( angles-anglesRef  )
Ytournant = rho * np.sin( angles-anglesRef  )


plt.figure(figsize=(7,7))
ax=plt.subplot(aspect='equal')
ax.add_artist( plt.Circle((100*allX[0], 100*allY[0]), 100*R, alpha=0.2) )

ax.plot( 100*Xtournant, 100*Ytournant )
ax.plot( 100*allX[0], 100*allY[0], 'o' )


ax.plot( 0, 0, 'o' )
plt.xlabel('x tournant (cm)')
plt.ylabel('y tournant (cm)')
```
